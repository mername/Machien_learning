```python
df = pd.read_csv("insurance.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
def pie_plot(column):
    fig, ax = plt.subplots()
    ax.pie(df[column].value_counts(), autopct = "%0.2f%%", labels = df[column].value_counts().index)
    ax.set(title = f"pie chart of {column}")
```

```python
columns = ['sex', 'smoker', 'region']
for i in columns:
    pie_plot(i)
```

```python
sns.pairplot(df)
```

```python
df['sex'].unique()
```

```python
df.head()
```

```python
df['smoker'].unique()
```

```python
mapper_smoker = {"yes": 1, "no":0}
df['smoker'] = df['smoker'].map(mapper_smoker)
df.head()
```

```python
df['region'].unique()
```

```python
mapper_region = {"southwest": 0, "southeast":1, "northwest":2, "northeast":3}
df['region'] = df['region'].map(mapper_region)
df.head()
```

```python
x = df.drop("charges", axis = 1)
y = df['charges']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```
